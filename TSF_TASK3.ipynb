### Importing The Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

### Importing the Dataset

```python
dataset=pd.read_csv('SampleSuperstore.csv')
df=pd.DataFrame(dataset)
df.head(10)
```

### Understanding the Data

### Dropping Redundant Columns.

```python
dataset.drop(['Ship Mode','Country','Region','Category'],axis=1,inplace=True)
dataset.head()
```

```python
dataset.shape
```

### Data Visualisation

#### Boxplot

```python
sns.boxplot(x=dataset['Sales'])
```

```python
sns.boxplot(x=dataset['Discount'])
```

```python
sns.boxplot(x=dataset['Profit'])
```

#### Heatmap

```python
corelation=dataset.corr()
plt.figure(figsize=(10,7))
sns.heatmap(corelation,xticklabels=corelation.columns,yticklabels=corelation.columns,annot=True)
```

The above heatmap shows the corelation between features in the data.

#### Pairplot

```python
sns.pairplot(dataset)
```

```python
sns.distplot(dataset['Sales'])
```

```python
sns.distplot(dataset['Profit'])
```

```python
g=sns.displot(dataset,x='Sub-Category',hue='Quantity')
g.fig.set_figwidth(17.7)
```

#### Scatter plot

```python
sns.relplot(x='Sales',y='Profit',hue='Sub-Category',data=dataset)
```

```python
sns.relplot(x='Sales',y='Profit',hue='Segment',data=dataset)
```
